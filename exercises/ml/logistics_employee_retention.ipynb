```python
import pandas as pd
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("data\\HR_comma_sep.csv")
df.head()
```

```python
mean_values = df.groupby("left").mean()
mean_values
```

```python
lr_model = LogisticRegression(solver='lbfgs', max_iter=1000000000)
```

```python
df_inputs = df[['satisfaction_level', 'average_montly_hours', 'promotion_last_5years', 'salary']]
df_inputs.head()
```

```python
target = df['left']
```

```python
salary_dummies = pd.get_dummies(df_inputs['salary'], prefix='salary')
salary_dummies.head()
```

```python
df_final = pd.concat([df_inputs.drop(['salary'], axis='columns'),
                      salary_dummies.drop(['salary_medium'], axis='columns')
                     ],
                     axis='columns')
df_final.head()
```

```python
x_train, x_test, y_train, y_test = train_test_split(df_final, target, train_size=0.7)
```

```python
lr_model.fit(x_train, y_train)
```

```python
lr_model.score(x_test, y_test)
```
